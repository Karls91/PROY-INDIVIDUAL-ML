### Realizando el merge1

```python
import pandas as pd
```

```python
items = pd.read_csv(r'C:\Users\Coder\Documents\PI_ML_OPS\Datos\items1.csv')
games = pd.read_csv(r'C:\Users\Coder\Documents\PI_ML_OPS\Datos\games1.csv')
```

Realizamos el Merge de ambos dataframes:

```python
merge_games_items =(pd.merge(items, games, left_on="item_id", right_on="item_id")[["user_id","item_id","item_name","publisher","genres","title", "release_date", "price","playtime_forever","developer"]])
```
