```python
caracteristicas = ['user_id','title','playtime_forever','developer','genres','price','posted','item_id','recommend','sentiment_analysis', 'release_date']
filtro = df_ml[caracteristicas]
filtro
```

```python
filtro[caracteristicas] = filtro[caracteristicas].astype(str)
```

```python
for caracteristica in caracteristicas:
    filtro[caracteristica]=filtro[caracteristica].apply(lambda x: x.lower().replace(" ",""))
```

```python
filtro.shape
```

```python
def crear_sopa(dato):
    return dato['user_id']+ " "+ dato['title']+ " " + " " + dato['playtime_forever']+ " " + dato['developer']+ " " + dato['genres']+ " "+ dato['price']+ " "+ dato['release_date']+ " "+ dato['posted']+ " "+ dato['item_id']+ " "+ dato['recommend']+ " "+ dato['sentiment_analysis']
```

```python

filtro['caracteristicas']=filtro.apply(crear_sopa, axis=1)
```

```python
filtro
```

```python
filtro[caracteristicas]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

modelo=TfidfVectorizer(stop_words='english')
matriz_modelo = modelo.fit_transform(filtro[caracteristicas])
```

```python
modelo
```

```python
matriz_modelo
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similitudCoseno=cosine_similarity(X=matriz_modelo, Y=matriz_modelo)
```

```python

similitudCoseno.shape
```

```python
similitudCoseno= pd.DataFrame(similitudCoseno)
similitudCoseno
```

```python
filtro = filtro.reset_index()
indice = pd.Series(filtro.index, index=filtro['item_id'])
```

```python
def obtener_recomendacion(item_id):
    idx = indice[item_id]
    print(idx)
    puntajes_similares = list(enumerate(similitudCoseno[idx]))
    puntajes_similares = sorted(puntajes_similares, key=lambda x: x[1], reverse=True)
    puntajes_similares = puntajes_similares[1:6]
    juegos_indices = [int(i[0]) for i in puntajes_similares]
    return {f'Recomendación de 5 juegos relacionados con ({item_id})' +df_ml['title'][item_id]:[{f'Recomendacion {i+1}':{'Identificador':df_ml['item_id'].iloc[juegos_indices[i]], 'Titulo':df_ml['title'].iloc[juegos_indices[i]]}} for i in range(len(juegos_indices))]}
```

```python
item_id = 7
obtener_recomendacion(item_id)
```
